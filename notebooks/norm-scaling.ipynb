```python
xs = np.linspace(0, 1, 100)
ys = 1 - xs
print(xs[:10], ys[:10])
```

```python
Es = []
for x, y in zip(xs, ys):
    vec = [x, y]
    Es.append(np.linalg.norm(vec))
Es = np.asarray(Es)
```

```python
plt.plot(xs, Es, color="black")
plt.xlabel("X")
plt.ylabel("E")
```
