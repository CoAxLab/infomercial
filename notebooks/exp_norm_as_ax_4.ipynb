```python
from copy import deepcopy
from random import shuffle

import numpy as np
from scipy.stats import entropy

from IPython.display import Image
import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'retina'

import seaborn as sns
sns.set_style('ticks')

matplotlib.rcParams.update({'font.size': 16})
matplotlib.rc('axes', titlesize=16)
```

```python
# Create a vector
x = np.random.randint(1, 1000, size=3)
x = x / x.sum()
print(x)

# Perturb is locally or globally
glob = []
loca = []
deltas = np.linspace(0, 1 - x[0], 100)
for delta in deltas:
    y1 = x.copy()
    y1[0] += delta
    y1 /= y1.sum()
    glob.append(np.linalg.norm(y1))
    
    y2 = x.copy()
    y2[0] += delta
    y2[1] -= delta
    loca.append(np.linalg.norm(y2))
    
# Plot
fig = plt.figure(figsize=(7, 4))
grid = plt.GridSpec(1, 1, wspace=0.3, hspace=0.8)
plt.subplot(grid[0, 0])
plt.plot(deltas, glob, color="grey", alpha=0.6, linewidth=6, label="G")
plt.plot(deltas, loca, color="goldenrod", alpha=0.6, linewidth=6, label="L")
plt.ylabel("Norm")
plt.xlabel("Delta")
plt.legend()
```
